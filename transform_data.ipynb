```python
import pandas as pd

data = {
    'benzinske': pd.read_csv('./data/benzinske.csv').loc[:, ['X', 'Y', 'NAZIV']],
    'wifi': pd.read_csv('./data/besplatni_wifi.csv').loc[:, ['X', 'Y', 'Lokacija']],
    'javni': pd.read_csv('./data/javni_prijevoz.csv').loc[:, ['X', 'Y', 'Naziv_stajališta']],
    'sport': pd.read_csv('./data/sport.csv').loc[:, ['X', 'Y', 'naziv']],
    'vrtici': pd.read_csv('./data/vrtici.csv').loc[:, ['X', 'Y', 'naziv']]
}

data_global = pd.DataFrame(columns=['X', 'Y', 'naziv', 'class'])

for d in data:
    data[d].columns = ['X', 'Y', 'naziv']
    data[d]['class'] = [d]*len(data[d])

    print(data[d])

    data_global = pd.concat([data_global, data[d]])

data_global
```

```python
data_global.to_csv('data.csv', index=False)
```
